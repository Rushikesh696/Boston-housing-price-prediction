### Outlier Detection:
- After analyzing the numerical columns, it was found that there are not many outliers present.
- The crim, zn, b feature have the highest number of outliers, totaling 60-80.
- The remaining numerical columns each contain fewer than 40 outliers.

```python
for i in boston.columns[:-1]:
    if boston[i].dtype==object:
        continue
    else:
        q1=boston[i].quantile(0.25)
        q3=boston[i].quantile(0.75)
        IQR=q3-q1
        lower_limit=q1-(1.5*IQR)
        upper_limit=q3+(1.5*IQR)
        print(i,  " :" ,len(boston[(boston[i]<lower_limit) | (boston[i]>upper_limit)]))
        boston.loc[(boston[i]<lower_limit) | (boston[i]>upper_limit),i]=boston[i].median()
        print(i, " :"  ,len(boston[(boston[i]<lower_limit) | (boston[i]>upper_limit)]))
```

## simple linear regression
- Simple linear regression is a statistical method used to model the relationship between two variables by fitting a straight line to the data.

- The simple linear regression model shows moderate performance with an R value of 0.54 on training and 0.55 on testing, meaning it explains about 54-55% of the variance in medv based on lstat.
- Ridge regression yields similar results, suggesting regularization doesn't have a significant impact here.
- Lasso regression slightly improves performance, with a testing R of 0.55, indicating better generalization.
- ElasticNet regression, combining Lasso and Ridge, performs the best, with an R of 0.55 on both training and testing sets, highlighting its ability to capture the relationship more effectively.
- Overall, the models with regularization (Lasso and ElasticNet) show a slight edge in performance over simple linear regression.

```python
# Ridge
input_variable=boston["lstat"].values.reshape(-1,1)
output_variable=boston["medv"].values
input_train, input_test, output_train, output_test=train_test_split(input_variable, output_variable, test_size=0.2,random_state=42)
ridge_reg=Ridge()
ridge_reg.fit(input_train, output_train)
# for training
y_pred_ridge_train=ridge_reg.predict(input_train)
# for testing 
y_pred_ridge_test=ridge_reg.predict(input_test)
print(f' RIDGE regression TRAINING  :',ridge_reg.score(input_train, output_train))
print("*"*25)
print(f' RIDGE regression TESTING   :',ridge_reg.score(input_test, output_test))
```

```python
# lasso
input_variable=boston["lstat"].values.reshape(-1,1)
output_variable=boston["medv"].values

input_train, input_test, output_train, output_test=train_test_split(input_variable, output_variable, test_size=0.2,random_state=42)

lasso_reg=Lasso()
lasso_reg.fit(input_train, output_train)
# for training
y_pred_lasso_train=lasso_reg.predict(input_train)
# for testing 
y_pred_lasso_test=lasso_reg.predict(input_test)
print(f' LASSO regression TRAINING  :',lasso_reg.score(input_train, output_train))
print("*"*25)
print(f' LASSO regression TESTING   :',lasso_reg.score(input_test, output_test))
```

```python
# elasticnet
input_variable=boston["lstat"].values.reshape(-1,1)
output_variable=boston["medv"].values

input_train, input_test, output_train, output_test=train_test_split(input_variable, output_variable, test_size=0.2,random_state=42)

elasticnet_reg=ElasticNet()
elasticnet_reg.fit(input_train, output_train)
# for training
y_pred_elasticnet_train=elasticnet_reg.predict(input_train)
# for testing 
y_pred_elasticnet_test=elasticnet_reg.predict(input_test)
print(f' ELASTICNET regression TRAINING :',elasticnet_reg.score(input_train, output_train))
print("*"*25)
print(f' ELASTICNET regression TESTING :',elasticnet_reg.score(input_test, output_test))
```

## Polynomial regression
- Polynomial regression is typically used when there is non-linearity in the data and the data points follow a curved pattern.
- The goal of polynomial regression is to find a curve that fits the data points as closely as possible, minimizing errors and providing a more accurate representation of the underlying pattern.
- By using a curve, the model can capture the complexity of the data and deliver predictions with lower error compared to a straight line.

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
x=boston["lstat"].values.reshape(-1,1)
y=boston["medv"]
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.2, random_state=40)
poly=PolynomialFeatures(degree=2)
xtrainpoly=poly.fit_transform(x_train)
xtestpoly=poly.fit_transform(x_test)
lr=LinearRegression()
lr.fit(xtrainpoly,y_train )
```

```python
trainpoly=lr.predict(xtrainpoly)
testpoly=lr.predict(xtestpoly)
```

```python
r2_score(y_test, testpoly)
```

```python
r2_score(y_train, trainpoly)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(x_train, y_train, color='blue', label='Train data')
x_sorted = np.sort(x_train, axis=0)
x_sorted_poly = poly.transform(x_sorted)
train_curve = lr.predict(x_sorted_poly)
plt.plot(x_sorted, train_curve)
```

```python
plt.scatter(x_test, y_test)
plt.plot(x_sorted, train_curve)
```

### Residual Plot
- Residuals represent the difference between the actual values and the predicted values from the model.
- They help us assess how well the model fits the data.
- Residuals measure the error or deviation between the actual observations and the models predictions.
- Smaller residuals indicate that the model is making accurate predictions, meaning it fits the data well.
- Conversely, larger residuals suggest that the model is not performing well and may provide inaccurate predictions.

```python
residual=y_test-testpoly
sns.scatterplot(x=testpoly, y=residual)
plt.show()
```
